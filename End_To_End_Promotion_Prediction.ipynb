```python
train.select_dtypes('number').head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))

sns.boxplot(x=train['avg_training_score'], color='red', ax=ax[0])
ax[0].set_xlabel('Average Training Score')
ax[0].set_ylabel('Count')

sns.boxplot(x=train['length_of_service'], color='red', ax=ax[1])
ax[1].set_xlabel('Length of Service')
ax[1].set_ylabel('Count')

plt.suptitle('Outliers Detection')
plt.show()
```

We have outliers in Service chart

```python
Q1 = train['length_of_service'].quantile(0.25)
Q3 = train['length_of_service'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

train_no_outliers = train[(train['length_of_service'] >= lower_bound) & (train['length_of_service'] <= upper_bound)]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8))

sns.boxplot(x=train_no_outliers['avg_training_score'], color='green', ax=ax[0])
ax[0].set_xlabel('Average Training Score')
ax[0].set_ylabel('Count')

sns.boxplot(x=train_no_outliers['length_of_service'], color='green', ax=ax[1])
ax[1].set_xlabel('Length of Service')
ax[1].set_ylabel('Count')

plt.suptitle('Outliers Detection')
plt.show()
```

# Handling Imbalanced Data

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X, y = smote.fit_resample(X, y)
```

# DecisionTree Classifier

```python
from sklearn.metrics import accuracy_score, roc_curve, auc
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_val)
```

```python
print('Trainning Accuracy: ', model.score(X_train, y_train))
print('Testing Accuracy: ', model.score(X_val, y_val))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

conm_dt = confusion_matrix(y_val, y_pred)
sns.heatmap(conm_dt, annot=True, fmt='d')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix of Decision Tree Classifier')
plt.show()
```

```python
evaluation = classification_report(y_val, y_pred)
print(evaluation)
```

```python
# Get predicted probabilities for ROC curve
y_pred_prob_val = model.predict_proba(X_val)[:, 1]  # Probability of the positive class

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_val, y_pred_prob_val)

# Calculate the AUC (Area Under the Curve)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.1, 1.0])
plt.ylim([-0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve of Decision Tree Classifier')
plt.legend(loc="lower right")
plt.show()
```
